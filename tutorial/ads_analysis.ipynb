```python
from actgpaw import utils as ut
from ase.io import read
import glob
import matplotlib.pyplot as plt
import numpy as np
from ase.visualize.plot import plot_atoms
```

```python
element='Cu_mp-30'
surf_struc=['111']
ut.create_ads_and_dir(element,surf_struc,ads_atom=['Li'],ads_site=['ontop','hollow','bridge'])
```

```python
for struc in surf_struc:
    fil_loc=element+'/ads/'+struc+'/adsorbates/Li/*/*/*.traj'
    fil=glob.glob(fil_loc,recursive=False)
    fig = plt.figure(figsize=(15,15))
    for n,fil_name in enumerate(fil):
        ax=fig.add_subplot(np.ceil(len(fil)/2),2,n+1)
        plot_atoms(read(fil_name),ax)
        locations=fil_name.split('/')[-2]
        location_x=locations.split('_')[0]
        location_y=locations.split('_')[1]
        adsorbate=fil_name.split('/')[-4]
        ax.set_title('{}[{}]({},{})'.format(struc,adsorbate,location_x,location_y))
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
```
