```python
import numpy as np
# specify parameters
element = 'Cu_mp-30'
ind = (1,1,1)
option = 'pymatgen'
plot = False
order = 0
save = True

# specify the layer parameters
layers = np.array([4,6,8,10,12,14])
vacuum_layers = layers/2
for i in range(len(layers)):
    surf.surf_creator(element, ind, layers[i], vacuum_layers[i], 
                        unit=True, # if using pymatgen, specify the unit of layers (number[True]/length[False])
                        option=option,
                        save=save,
                        order=order,
                        plot=plot)
```
